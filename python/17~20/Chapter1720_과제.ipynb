과제_1: BungeoppangShop 클래스 만들기

- __init__: 초기 재고, 가격, 판매량을 설정합니다.
- check_stock: 현재 재고를 출력합니다.
- process_order: 주문을 처리하고 재고를 업데이트합니다.
- admin_mode: 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales: 총 매출을 계산하고 출력합니다.

```python
class BungeoppangShop:
    def __init__(self):
        self.stock={"팥붕어빵":10,"슈크림붕어빵":8,"초코붕어빵":5}
        self.prices={"팥붕어빵":1000,"슈크림붕어빵":1200,"초코붕어빵":1500}
        self.sales={"팥붕어빵":0,"슈크림붕어빵":0,"초코붕어빵":0}
        
    def check_stock(self):
        for key,value in self.stock.items():
            print(f"{key}의 재고는 {value}개 입니다.")

    def process_order(self, bread_type, bread_count):
        if bread_type in self.stock:
            if bread_count<=self.stock[bread_type]:
                self.stock[bread_type]-=bread_count
                self.sales[bread_type]+=bread_count
            else:
                print("재고가 부족합니다. 다시 주문해주세요.")
        else:
            print("잘못 입력하셨습니다.")

    def admin_mode(self,bread_type,additional_stock):
        if bread_type in self.stock:
            try:
                self.stock[bread_type]+=additional_stock
            except ValueError:
                print("값을 잘못 입력하였습니다.")
            except TypeError:
                print("숫자로 입력해주세요.")
        else:
            print("잘못 입력하였습니다.")

    def calculate_total_sales(self):
        total=0
        for (bread_type,sale),(bread_type2,price) in zip(self.sales.items(), self.prices.items()):
            total+=sale*price
            print(f"금일 매출은 {total}원입니다.")
```

[과제 4: 주문 처리 메서드 process_order]

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
def process_order(self, bread_type, bread_count):
    if bread_type in self.stock:
        if bread_count<=self.stock[bread_type]:
            self.stock[bread_type]-=bread_count
            self.sales[bread_type]+=bread_count
        else:
            print("재고가 부족합니다. 다시 주문해주세요.")
    else:
        print("잘못 입력하셨습니다.")
```

[과제 8: 메인 함수 main]

목표: main 함수를 작성하여 프로그램의 주요 로직을 실행합니다.

설명:
- BungeoppangShop 객체를 생성합니다.
- 사용자에게 모드를 선택하도록 요청합니다.
- "주문" 모드에서는 손님의 주문을 처리하고, "관리자" 모드에서는 재고를 추가할 수 있습니다.
- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
class BungeoppangShop:
    def __init__(self):
        self.stock={"팥붕어빵":10,"슈크림붕어빵":8,"초코붕어빵":5}
        self.prices={"팥붕어빵":1000,"슈크림붕어빵":1200,"초코붕어빵":1500}
        self.sales={"팥붕어빵":0,"슈크림붕어빵":0,"초코붕어빵":0}
        
    def check_stock(self):
        for key,value in self.stock.items():
            print(f"{key}의 재고는 {value}개 입니다.")

    def process_order(self, bread_type, bread_count):
        if bread_type in self.stock:
            if bread_count<=self.stock[bread_type]:
                self.stock[bread_type]-=bread_count
                self.sales[bread_type]+=bread_count
                print(f"{bread_type}이 {bread_count}개 팔렸습니다.")
            else:
                print("재고가 부족합니다. 다시 주문해주세요.")
        else:
            print("잘못 입력하셨습니다.")

    def admin_mode(self,bread_type,additional_stock):
        if bread_type in self.stock:
            try:
                self.stock[bread_type]+=additional_stock
                print("\n")
                print(f"{bread_type}이 {additional_stock}개 추가되었습니다.")
            except ValueError:
                print("값을 잘못 입력하였습니다.")
            except TypeError:
                print("숫자로 입력해주세요.")
                
        else:
            print("잘못 입력하였습니다.")

    def calculate_total_sales(self):
        total=0
        for (bread_type,sale),(bread_type2,price) in zip(self.sales.items(), self.prices.items()):
            total+=sale*price
            print(f"금일 매출은 {total}원입니다.")
            
def main():
    shop=BungeoppangShop()
    while True:
        mode=input("모드를 선택해주세요 (주문/관리자/종료) : ")
        if mode=="주문":
            taste=input("어떤 붕어빵을 주문하시겠어요?")
            try:
                개수=int(input(f"{taste} 몇 개를 구매하시겠어요?"))
                shop.process_order(taste,개수)
            except ValueError:
                print("정수로 입력해주세요.")
                continue
            except TypeError:
                print("잘못 입력하셨습니다.")
                continue

        elif mode=="관리자":
            print("=============재고=============")
            shop.check_stock()
            add_taste=input("어떤 붕어빵을 추가하시겠어요?")
            try:
                add_개수=int(input(f"{add_taste} 몇 개를 추가하시겠어요?"))
                shop.admin_mode(add_taste,add_개수)
            except ValueError:
                print("정수로 입력해주세요.")
                continue
            except TypeError:
                print("잘못 입력하셨습니다.")
                continue

        elif mode=="종료":
            print("=============재고=============")
            shop.check_stock()
            print("=============매출=============")
            shop.calculate_total_sales()
            break
            
        else:
            print("잘못 입력하셨습니다.")
            continue
                
if __name__ == "__main__":
    main()
```
